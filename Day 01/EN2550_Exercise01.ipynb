```python
def HyEst(S):
    a = S
    n =0
    while not (1<=a<100):
        a = int(S//100)
        n+=1
    return ((-190/(a+20))+10)*(10**n)
print(HyEst(1600))
```

```python
def Sqrt(N):
    sq_p = HyEst(N)
    sq_n =0
    
    while True:
        sq_n = 0.5*(sq_p+N/sq_p)
        if (-10e-5)<(sq_n-sq_p)<(10e-5):
            break
        sq_p = sq_n
    return round(sq_n,6)

print("Sqrt of 64:",Sqrt(64))
print("Sqrt of 75:",Sqrt(75))
print("Sqrt of 100:",Sqrt(100))
print("Sqrt of 1600:",Sqrt(1600))
```
