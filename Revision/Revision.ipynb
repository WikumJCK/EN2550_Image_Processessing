### __02) Opening and displaying images__

```python
#Using Matplotlib to open an image

img = cv.imread(r'./Images/spider.png',cv.IMREAD_COLOR)
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

fig , ax = plt.subplots(facecolor = 'white')
ax.imshow(img)
ax.set_title('Image opened in Matplotlib')
ax.axis('off')
plt.show()
```

### __03) Spatial filtering an Image__

```python
#Averaging Filter (This will cause a blur Effect in the image)
img = cv.imread(r'./Images/spider.png',cv.IMREAD_GRAYSCALE)

kernal = np.ones((3,3),np.float32)/9
imgc = cv.filter2D(img,-1,kernal)

fig , ax = plt.subplots(3,1,sharex='all',sharey='all',figsize =(18,18),facecolor = 'white')
ax[0].imshow(img,cmap ='gray')
ax[0].set_title('Original')
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].imshow(imgc,cmap ='gray')
ax[1].set_title('Averaging')
ax[1].set_xticks([])
ax[1].set_yticks([])

img_f = img+(img-imgc)
ax[2].imshow(img_f,cmap ='gray')
ax[2].set_title('Reduced')
ax[2].set_xticks([])
ax[2].set_yticks([])

plt.show()
```

```python
#Spatial Filtering (Sobel Verticle and Soble horizontal Filtering)
img = cv.imread(r'./Images/spider.png',cv.IMREAD_GRAYSCALE)

kernal_v = np.array([(-1,-2,-1),(0,0,0),(1,2,1)],dtype = 'float')
img_v = cv.filter2D(img,-1,kernal_v)

kernal_h = np.array([(-1,0,1),(-2,0,2),(-1,0,1)],dtype = 'float')
img_h = cv.filter2D(img,-1,kernal_h)

fig , ax = plt.subplots(4,1,sharex='all',sharey='all',figsize =(18,18),facecolor ='white')
ax[0].imshow(img,cmap ='gray')
ax[0].set_title('Original')
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].imshow(img_v,cmap ='gray')
ax[1].set_title('Soble Verticle')
ax[1].set_xticks([])
ax[1].set_yticks([])


ax[2].imshow(img_h,cmap ='gray')
ax[2].set_title('Soble Horizontal')
ax[2].set_xticks([])
ax[2].set_yticks([])

imgx = img-(img_h+img_v)
ax[3].imshow(imgx,cmap ='gray')
ax[3].set_title('Soble Horizontal')
ax[3].set_xticks([])
ax[3].set_yticks([])
plt.show()
```

### __04) Partial derivative of an Image__

```python
img = cv.imread(r'./Images/spider.png',cv.IMREAD_GRAYSCALE)

kernal_h = np.array([(-1,-1,-1),(0,0,0),(1,1,1)],dtype = 'float')
img_h = cv.filter2D(img,-1,kernal_h)

kernal_v = np.array([(-1,0,1),(-1,0,1),(-1,0,1)],dtype = 'float')
img_v = cv.filter2D(img,-1,kernal_v)

fig , ax = plt.subplots(3,1,sharex='all',sharey='all',figsize =(18,18),facecolor ='white')
ax[0].imshow(img,cmap ='gray')
ax[0].set_title('Original')
ax[0].set_xticks([])
ax[0].set_yticks([])

ax[1].imshow(img_v,cmap ='gray')
ax[1].set_title('Verticle')
ax[1].set_xticks([])
ax[1].set_yticks([])

ax[2].imshow(img_h,cmap ='gray')
ax[2].set_title('Horizontal')
ax[2].set_xticks([])
ax[2].set_yticks([])
plt.show()
```
